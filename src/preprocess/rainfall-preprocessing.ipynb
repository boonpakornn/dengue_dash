```python
import os
import json
import pandas as pd
import numpy as np
from tqdm import tqdm_notebook

df_rainfall_sum = pd.read_csv(os.path.join('..', '..', 'data', 'weather', 'csv', 'rainfall-subdistrict-sum.csv'))
df_rainfall_avg = pd.read_csv(os.path.join('..', '..', 'data', 'weather', 'csv', 'rainfall-subdistrict-avg.csv'))
df_rainfall_sum.head(1)
df_rainfall_avg.head(1)
```

```python
months = df_rainfall_sum.columns[1:].values
addrcodes = df_rainfall_sum['TB_IDN'].values
```

```python
indexes_addrcode = []
indexes_month = []

sum_values = []
avg_values = []

for addrcode in addrcodes:
    for month in months:
        indexes_addrcode.append(addrcode)
        indexes_month.append(month)
        
    sum_values += list(df_rainfall_sum.loc[df_rainfall_sum['TB_IDN'] == addrcode].values[0][1:])
    avg_values += list(df_rainfall_avg.loc[df_rainfall_avg['TB_IDN'] == addrcode].values[0][1:])
        
indexes_addrcode = np.reshape(np.array(indexes_addrcode),(-1, 1))
indexes_month = np.reshape(np.array(indexes_month),(-1, 1))

sum_values = np.reshape(np.array(sum_values),(-1, 1))
avg_values = np.reshape(np.array(avg_values),(-1, 1))
rainfall_values = np.concatenate((indexes_addrcode, indexes_month, sum_values, avg_values), axis=1)
```

```python
df_rainfall = pd.DataFrame.from_records(rainfall_values, columns=['addrcode','date','rainfall_sum','rainfall_avg'])
df_rainfall['date'] = pd.to_datetime(df_rainfall['date'], format='%Y_%m')
df_rainfall.set_index(['addrcode','date'], inplace=True)
df_rainfall.head(10)
```

```python
df_rainfall.to_csv(os.path.join('..', '..', 'data', 'weather', 'rainfall.csv'))
```
